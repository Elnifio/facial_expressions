```python
def readimage(readpath, storepath, extra):
    img = Image.open(readpath).convert("L")
    if (img.size[0] != img.size[1]):
        extra.append(readpath)
    else:
        img = img.resize((227, 227), Image.ANTIALIAS)
        img.save(storepath)
```

```python
ls = os.listdir("images")
```

```python
extra = []

for item in ls:
    readimage("./images/" + item, "./grayscale/"+item, extra)
    print(".", end="")
```
